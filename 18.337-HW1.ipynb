## Flipping Coins

```julia
# We can establish definitions on all processors by surrounding a cell in `@everywhere`.
# This definition is so small that it's not really necessary, but we provide it as an example.
@everywhere begin
    coinflip() = int(randbool())
end
```

```julia
# monte carlo estimate of probability of heads
n = 10000000
nheads = @parallel (+) for i=1:n; coinflip(); end
nheads/n
```

## Problem 1. Longest Increasing Subsequence

```julia
# length of longest increasing subsequence of a vector
@everywhere function patiencesort(p::Vector)
    piles = similar(p, 0)
    for pi in p
        idx = 1
        for pp in piles
            if pi > pp
                idx += 1
            end
        end
        if idx > length(piles)
            d = length(piles)
            resize!(piles, idx)
            piles[d+1:end] = 0
        end
        piles[idx] = pi
    end
    return length(piles)
end
```
