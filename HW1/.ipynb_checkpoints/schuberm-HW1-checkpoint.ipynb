## Flipping Coins

```julia
# We can establish definitions on all processors by surrounding a cell in `@everywhere`.
# This definition is so small that it's not really necessary, but we provide it as an example.
@everywhere begin
    coinflip() = int(randbool())
end
```

```julia
# monte carlo estimate of probability of heads
n = 10000000
nheads = @parallel (+) for i=1:n; coinflip(); end
nheads/n
```

## Problem 1. Longest Increasing Subsequence

```julia
# length of longest increasing subsequence of a vector
@everywhere function patiencesort(p::Vector)
    piles = similar(p, 0)
    for pi in p
        idx = 1
        for pp in piles
            if pi > pp
                idx += 1
            end
        end
        if idx > length(piles)
            d = length(piles)
            resize!(piles, idx)
            piles[d+1:end] = 0
        end
        piles[idx] = pi
    end
    return length(piles)
end
```

```julia
# Histogram (in parallel) the longest increasing subsequences of some big random
# permutations (see Julia function `randperm`).
# The resulting distribution is understood, but not widely known.

# Note that the value of a `@parallel` loop body can be anything, and the reduction
# function can also be any (associative) function.

#Using pmap
n = 1000
bins = 1:1:floor(sqrt(n))*4
_,h = hist(pmap(patiencesort, {randperm(n) for i=1:n}),bins)
plot(bins[1:end-1], h)
```

```julia
#Using @parallel
seqs=@parallel (vcat) for i=1:n; patiencesort(randperm(n)); end
bins = 1:1:floor(sqrt(n))*4
_,h = hist(seqs,bins)
plot(bins[1:end-1], h)
```

## Problem 2. Generalize That

```julia
# Design and implement a utility function that makes it easier to run
# parallel monte carlo experiments, varying key parameters such as the
# number of trials.
```

```julia
@everywhere function gen_mc(f,n,op)
    seqs = @parallel (op) for i=1:n; f(); end
    return seqs/n
end

#overload to allow f to take arguments
#is there a better way?
@everywhere function gen_mc(f,p,n,op)
    seqs = @parallel (op) for i=1:n; f(randperm(p)); end
    return seqs
end
```

```julia
gen_mc(patiencesort,100,1000,vcat)
```

```julia
gen_mc(coinflip,1000,+)
```

## Problem 3. Your Experiment

```julia
#find the length of the zig zag of a random vector
# b is a vector of sign changes of lenghth length(p)-1
# 1 3 6 2 5 4 8 9 converts to 1 1 0 1 0 1 1 so the lengt is length[ 1 0 1 0 1]=5

function lacs(p::Vector)
    b = zeros(length(p))-1
    a = 0
    for i=1:length(p)-1
        if (p[i]-p[i+1])>=0
            b[i]=0
        else
            b[i]=1
        end
    end
    #return b
    for i=1:length(b)-1
        if (b[i]!=b[i+1])
            a+=1
        else
        end
    end
    return a
end
```

```julia
lacs([1,3,6,2,5,4,8,9])
```

```julia
n = 10000
l = 10000
bins = 1:1:n
_,h = hist(gen_mc(lacs,l,n,vcat),bins)
plot(bins[1:end-1], h)
```
